## Merge Sort 

```python
def merge_sort(numbers):
    # On veut diviser la liste en deux par récurrence jusqu'à n'avoir qu'un seul élément
    if len(numbers) < 2 : return numbers

    M = len(numbers)//2 #index du centre pour séparer en deux
    left = merge_sort(numbers[:M]) # M non compris
    right = merge_sort(numbers[M:]) # M compris
    left.reverse()
    right.reverse()
    # Rajouter une étape pour fusionner les listes séparés une fois qu'on a l'ordre
    sorted_numbers = []
    while left and right : # donc si on a les deux list, s'éxécute après les séparation left/right
        sorted_numbers += [left.pop() if left[-1] > right[-1] else right.pop()]
        
        # pop retire l'élément, while se break une fois la liste left ou right vide

    print("left: ", left, "\n right: ", right)
    return sorted_numbers + left + right # left et right parce que l'une des deux seuelement est vide après la boucle !


cards = [3, 7, 4, 1, 2, 7, 3]
sorted_cards = merge_sort(cards)

print(sorted_cards)


# Complexité de chaque ligne de la fonction merge_sort:

# def merge_sort(numbers):                                    # O(1) - définition de fonction
#     if len(numbers) < 2 : return numbers                    # O(1) - comparaison et retour
#     M = len(numbers)//2                                     # O(1) - division entière
#     left = merge_sort(numbers[:M])                          # O(n log n) - appel récursif sur moitié gauche
#     right = merge_sort(numbers[M:])                         # O(n log n) - appel récursif sur moitié droite
#     sorted_numbers = []                                     # O(1) - initialisation liste vide
#     while left and right :                                  # O(n) - boucle sur tous les éléments
#         sorted_numbers += [left.pop(0) if left[0] < right[0] else right.pop(0)]  # O(n) - pop et comparaison
#     return sorted_numbers + left + right                    # O(n) - concaténation des listes

# Complexité globale de merge_sort: O(n log n)
```
